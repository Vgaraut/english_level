```python
def optimize(source, idioms):
    path = source
    files = os.listdir(path)
    d=[]
    for file in files:
        filename, file_extension = os.path.splitext(file)
        cleaned = []
        total = []
        with open(os.path.join(path, file)) as s:
            sub = s.readlines()
        sub = list(map(lambda x: x.strip(), sub)) 
        for i in idioms:
            total.append(process.extractOne(i, sub, scorer=fuzz.token_sort_ratio))
        for idiom in total:
            if idiom[1] > 75:
                cleaned.append(idiom)
        d.append(len(cleaned))
    return d
```
